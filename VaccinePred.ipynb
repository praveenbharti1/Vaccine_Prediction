# Domain Analysis

#### What is H1N1 Virus ?
- H1N1 virus is a subtype of influenza A virus. It is a respiratory illness that can cause mild to severe illness in people of all ages. The virus is spread through respiratory droplets produced when an infected person coughs or sneezes.
    
#### what is Vaccine ?
- The H1N1 vaccine is a vaccine that helps protect against the H1N1 virus. The vaccine is made from a weakened or killed form of the virus. When you get the vaccine, your body will develop antibodies to the virus. These antibodies will help protect you from getting sick if you are exposed to the virus.
- The H1N1 vaccine is available in two forms:

    - Inactivated vaccine (the shot): This vaccine contains a killed form of the virus. It is injected into the muscle.
    - Live attenuated vaccine (the nasal spray): This vaccine contains a weakened form of the virus. It is sprayed into the nose.
